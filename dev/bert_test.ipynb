```python
# Testing BERT encoding
from sentence_transformers import SentenceTransformer

# Load a pre-trained BERT model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Assuming 'corpus' is a list of text documents
corpus_embeddings = model.encode(['This is a test sentence for BERT'])
```

```python
from sentence_transformers import SentenceTransformer, util

tokenizer = model.tokenizer

# Sentence to encode
sentence = "This is a test sentence for BERT! Come as you are."

# Tokenize the sentence
tokens = tokenizer.tokenize(sentence)

print("Tokens:", tokens)
```

```python
print("Embedding:", corpus_embeddings)
```
