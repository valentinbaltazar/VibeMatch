```python
# Pre trained BERT model testing

import pandas as pd

df0 = pd.read_csv('./data/clean_lyrics.csv')

df0.head()
```

```python
df = df0.dropna(subset=['corpus'],ignore_index=True)
```

```python
df.info()
```

```python
artists_to_keep = ['HAIM','Joyner Lucas','Phil Wickham']
corpus = df[['artist','title','corpus']][df['artist'].isin(artists_to_keep)].reset_index()
```

```python
corpus.info()
```

```python
corpus.head()
```

```python
from sklearn.cluster import KMeans
from sentence_transformers import SentenceTransformer
import numpy as np

# Load a pre-trained BERT model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Assuming 'corpus' is a list of text documents
corpus_embeddings = model.encode(corpus['corpus'])
```

```python

# Define the number of clusters
num_clusters = 3

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters,random_state=42)
kmeans.fit(corpus_embeddings)
clusters = kmeans.predict(corpus_embeddings)

# Printing cluster results
for cluster_id in range(num_clusters):
    print(f"Cluster {cluster_id}:")
    for i, name in enumerate(corpus['artist']):
        if clusters[i] == cluster_id:
            print(name,corpus['title'][i],cluster_id)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Assuming 'corpus' is the original list of text documents
# 'clusters' is the array of cluster labels obtained from K-means clustering
# 'num_clusters' is the number of clusters

# Create a dictionary to store texts for each cluster
cluster_texts = {i: [] for i in range(num_clusters)}

# Populate the dictionary with texts belonging to each cluster
for i, text in enumerate(corpus['corpus']):
    cluster_texts[clusters[i]].append(text)

# Generate WordCloud for each cluster
for cluster_id, texts in cluster_texts.items():
    # Combine all texts in the cluster into a single string
    cluster_text = " ".join(texts)
    
    # Generate the WordCloud
    wordcloud = WordCloud(width=800, height=400).generate(cluster_text)
    
    # Display the WordCloud
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f"Cluster {cluster_id} WordCloud")
    plt.axis('off')
    plt.show()
```
